```python
# initial button
button_score = widgets.Button(
    description = 'start diagnose',
    button_style = 'primary',
    # layout = layout_button
)

def button_score_click(_):
    global button_score_flag
    button_score_flag = True

button_score_flag = False # initial
button_score.on_click(button_score_click)
```

```python
#score wrapper
score = widgets.VBox(
    children = (score_age, score_BMI, score_glucose, score_insulin, score_HOMA, score_leptin, score_adiponectin, score_resistin, score_MCP, button_score), layout=Layout(
    display='flex',
    flex_flow='column',
    border='solid 2px',
    align_items='stretch',
    ))
```

```python
def predict_medicalrecord(score, model):
    '''
    score: [age, BMI, glucose, insulin, HOMA, leptin, adiponectin, resistin, MCP]
    '''

    result = model.predict([score])
    for prediction in result:
        if result == 1:
            prediction = 'Normal'
        else:
            prediction = 'Kanker'

  # prediction =  print(prediction)
    result = 'Result: ' + str(result) + ' ' + str(prediction)
    return result
```

```python
# label
label_current_thread = widgets.Label()
label_thread_list = widgets.Label()

flag_init_input = False


# button 
button_stop = widgets.Button(
    description = 'stop',
    icon = 'fa-stop',
    button_style = 'warning',
    layout = widgets.Layout(width='100px')
)
def button_stop_click(self):
    global thread_status
    thread_status = False
button_stop.on_click(button_stop_click)

button_check_thread = widgets.Button(
    description = 'check thread',
    button_style = '',
    layout = widgets.Layout(width='100px')
)
def button_check_thread_click(self):
    # print if work() still running
    work_count = 0
    name='work'
    for thread_object in threading.enumerate():
        if thread_object.name == name:
            work_count+=1
    if work_count > 0:
        label_thread_list.value=f'{work_count} thread with name {name} still exist. Consider use exit() and restart kernel'
    else:
        label_thread_list.value=f'No thread with name {name} exist'
button_check_thread.on_click(button_check_thread_click)
```

```python
# main loop
def work():

    global thread_status
    global flag_init_input 
    global model
    global button_score_flag

    
    if thread_status:
        label_current_thread.value = 'Thread On'

    current_state =[
        score_age.value,
        score_BMI.value,
        score_glucose.value,
        score_insulin.value,
        score_HOMA.value,
        score_leptin.value,
        score_adiponectin.value,
        score_resistin.value,
        score_MCP.value,
        ]
    
    while thread_status:
        # if button_score clicked
        if button_score_flag == True:

#         # check if there is a change in uploader.value
#         if current_state != [
#                              score_age.value,
#                              score_BMI.value,
#                              score_glucose.value,
#                              score_insulin.value,
#                              score_HOMA.value,
#                              score_leptin.value,
#                              score_adiponectin.value,
#                              score_resistin.value,
#                              score_MCP.value
#                              ]:
            
            # renew state reference
            current_state = [
                             score_age.value,
                             score_BMI.value,
                             score_glucose.value,
                             score_insulin.value,
                             score_HOMA.value,
                             score_leptin.value,
                             score_adiponectin.value,
                             score_resistin.value,
                             score_MCP.value
                             ]
            
            # # preprocess numpy
            # if score_age.value
            # if score_BMI.value
            # if score_glucose.value
            # if score_insulin.value
            # if score_HOMA.value
            # if score_leptin.value
            # if score_adiponectin.value
            # if score_resistin.value
            # if score_MCP.value

            # renew text_result.value
            text_result.value = predict_medicalrecord(current_state, model)

            if text_result.value:
                wrap_container.children = (score, text)
                
            button_score_flag = False
            
        if not thread_status:
            label_current_thread.value = 'Thread Off'
```

```python
# restart flag
thread_status = False  
thread_status = True

# create thread
thread_work = threading.Thread(target=work)
thread_work.name = 'work'

# start thread
thread_work.start()
```
